```python
import sys
from pathlib import Path

import cv2

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.face.verification import FaceVerifier, default_encoder_path
```

```python
GALLERY_PATH = ROOT / "dataset" / "faces_db"
ENCODER_PATH = default_encoder_path(ROOT)
COSINE_THRESHOLD = 0.42

verifier = FaceVerifier.load(
    gallery_root=GALLERY_PATH,
    encoder_path=ENCODER_PATH,
    cosine_threshold=COSINE_THRESHOLD,
    project_root=ROOT,
)
print("Registered:", verifier.identity_names)
```

```python
cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

cap = cv2.VideoCapture(0)
print("Press Q to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(80, 80))

    for x, y, w, h in faces:
        crop = frame[y : y + h, x : x + w]
        label, similarity = verifier.recognise_face(crop)
        color = (0, 255, 0) if label != "Unknown" else (0, 0, 255)
        cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
        cv2.putText(
            frame,
            f"{label} cos={similarity:.2f}",
            (x, max(20, y - 10)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            color,
            2,
        )

    cv2.imshow("Face Verification Test", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

cap.release()
cv2.destroyAllWindows()
```
